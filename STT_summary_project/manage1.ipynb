```python
import os
from moviepy.editor import VideoFileClip

# 비디오 파일이 있는 폴더 경로
video_folder_path = "./new_video/"

# 오디오 파일을 저장할 폴더 경로
audio_output_folder_path = "./new_audio/"

# 오디오 파일을 저장할 폴더가 없으면 생성
if not os.path.exists(audio_output_folder_path):
    os.makedirs(audio_output_folder_path)

# 비디오 폴더 내 모든 파일을 가져오기
for filename in os.listdir(video_folder_path):
    # 확장자가 비디오 파일인지 확인 (여기서는 .mp4 기준)
    if filename.endswith(".mp4"):
        video_path = os.path.join(video_folder_path, filename)
        audio_output_path = os.path.join(
            audio_output_folder_path, f"{os.path.splitext(filename)[0]}.mp3"
        )

        # 비디오 파일 불러오기
        video_clip = VideoFileClip(video_path)

        # 비디오에서 오디오 추출
        audio_clip = video_clip.audio

        # 오디오 파일로 저장
        print(f"Converting: {filename} -> {audio_output_path}")
        audio_clip.write_audiofile(audio_output_path)

        # 리소스 해제
        audio_clip.close()
        video_clip.close()

print("All video files have been converted to audio!")
```

```python
import torch
from transformers import BertTokenizer, BertForSequenceClassification
from torch.utils.data import DataLoader, Dataset
from sklearn.preprocessing import MultiLabelBinarizer

# 1. 카테고리 정의
classes = [
    "수",
    "덧셈",
    "뺄셈",
    "곱셈",
    "나눗셈",
    "분수",
    "소수",
    "소수점",
    "길이",
    "시간",
    "화폐",
    "측정",
    "각도",
    "평면도형",
    "입체도형",
    "도형",
    "대칭",
    "원",
    "직각",
    "비율",
    "비례식",
    "비",
    "그래프",
    "통계",
    "확률",
    "약수",
    "배수",
    "함수",
]

# 2. 토크나이저 및 모델 준비
tokenizer = BertTokenizer.from_pretrained("monologg/kobert")
model = BertForSequenceClassification.from_pretrained(
    "monologg/kobert", num_labels=len(classes)
)

# 3. 다중 레이블을 위한 데이터 전처리
mlb = MultiLabelBinarizer(classes=classes)


class MathDataset(Dataset):
    def __init__(self, texts, labels):
        self.texts = texts
        self.labels = mlb.fit_transform(labels)

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        inputs = tokenizer(
            self.texts[idx],
            padding="max_length",
            truncation=True,
            max_length=128,
            return_tensors="pt",
        )
        labels = torch.tensor(self.labels[idx], dtype=torch.float)
        return {**inputs, "labels": labels}


# 4. 학습 데이터 예시
texts = [
    "이 비디오를 잠시 멈추고, 실제로 27 더하기 1이 얼마인지, 가능하다면 27 더하기 10이 얼마인지 알아보세요. 그럼, 함께 생각해봅시다. 이 계산에는 크게 어려움이 없었을 것입니다. 1을 더한다는 것은 다음 숫자로 넘어가는 것이다, 그래서 27에서 한 번 더하면 28이 됩니다. 하지만 이 연습 문제를 하는 이유는 우리가 자릿수에 대해 고민하게 만듭니다. 여기서, 27이라는 수에서는, 2라는 수가 십의 자리에 있고, 이 2는 여기서 십의 자리입니다. 그리고 7은 일의 자리입니다. 그래서 27이라는 숫자를 두 개의 십과 일곱 개의 일로 표현할 수 있습니다. 그리고 이제 하나를 더한다는 건, 이 일의 자리 수를 하나 늘리는 것인데, 그렇다면 나는 얼마를 얻게 될까요? 여전히 두 개의 십이 있을 것이고, 그리고 얼마나 많은 일이 있을까요? 아, 여덟 개의 일이 있을 것입니다. 그러니까 결과는 28이 됩니다. 이것은 여러분에게 놀랍지 않을 수 있지만 자릿수에 대해 생각하는 것은 뒷받침되는 수학을 배울수록 매우 유용합니다. 이제 27 더하기 10에 대해 같은 것을 해봅시다. 우리는 이미 27을, 즉 두 십과 일곱 일로 말했었습니다. 그리고 이제 십에서, 일은 일의 자리가 아니라 더큰 수학을 배울수록 더 유용해집니다. 이제 27에 10을 더하면 어떻게 될까요? 우리는 이미 27, 즉 두 개의 십과 일곱 개의 일로 얘기했습니다. 그리고 이제 10에서, 1은 일의 자리에 있는 것이 아니며, 여기서의 1은 십의 자리입니다. 그래서 우리는 십 한 개와 일 없음을 가집니다. 그래서 십만이 그룹인 10이라는 숫자만 갖게 됩니다. 그래서 이 둘을 더하면 얼마가 될까요? 이제 여러분은 십을 세 개 가질 것입니다. 그래서 십의 자리에는 세 개가 됩니다. 그리고 여전히 일곱 개의 일을 가지고 있을 것입니다.  그러니까, 여기서는 1을 더하면 일의 자리에서 1만큼 증가한 것이고, 여기서는 10, 즉 십 한 개를 더하면 일의 자리가 늘어나지 않고 십의 자리가 늘어납니다.",
]
labels = [["덧셈", "수"], ["덧셈", "측정", "수"]]

# 5. 데이터셋 및 데이터로더 설정
dataset = MathDataset(texts, labels)
dataloader = DataLoader(dataset, batch_size=2)

# 6. 손실 함수 및 옵티마이저 정의
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-5)


# 7. 학습 함수 정의
def train_model(model, dataloader, optimizer, epochs=3):
    model.train()
    for epoch in range(epochs):
        for batch in dataloader:
            optimizer.zero_grad()
            outputs = model(
                input_ids=batch["input_ids"].squeeze(1),
                attention_mask=batch["attention_mask"].squeeze(1),
                labels=batch["labels"],
            )
            loss = outputs.loss
            loss.backward()
            optimizer.step()
        print(f"Epoch {epoch+1}/{epochs}, Loss: {loss.item()}")


# 모델 학습
train_model(model, dataloader, optimizer)
```
